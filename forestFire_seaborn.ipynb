```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
at = pd.read_csv(r"C:\Users\Shree\Desktop\atharva\DSBDA\datasets\forestfire.csv")
at
```

```python
sns.scatterplot(x='temp', y='area', data=at)
plt.title('Temperature vs. Area')
plt.xlabel('Temperature')
plt.ylabel('Area')
```

```python
sns.countplot(x='day', data=at, order=['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun'])
```

```python
sns.kdeplot(data=at['area'])
plt.title('Kernel Density Estimate of Fire Area')
plt.xlabel('Area')
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=at, x='month')
```

```python
sns.barplot(data=at, x='month', y='area', errorbar=None)
plt.title('Average Fire Area by Month')
plt.xlabel('Month')
plt.ylabel('Average Area')
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='month', y='wind', data=at)
plt.title('Wind Speed Distribution by Month')
plt.xlabel('Month')
plt.ylabel('Wind Speed')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='temp', data=at)
```

```python
plt.figure(figsize=(10, 8))
corr = at.corr()
sns.heatmap(corr, annot=True, cmap='YlGnBu')
```

```python
sns.lineplot(data=at, x='month', y='wind', estimator='mean')
plt.grid(True)
```
